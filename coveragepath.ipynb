```python
def distance(pointx1, pointx2,pointy1,pointy2):
        x_a= pointx1
        x_b = pointx2
        y_a= pointy1
        y_b= pointy2
        x_diff = x_a - x_b
        y_diff = y_a - y_b
        x_diff_sqr = x_diff**2
        y_diff_sqr = y_diff**2
        sumx_y = x_diff_sqr + y_diff_sqr
        final = math.sqrt(sumx_y)
        return final
```

```python
def calccost(pointx1, pointx2,pointy1,pointy2):
    px1=pointx1
    px2=pointx2
    py1=pointy1
    py2=pointy2
    distanceval=distance(px1,px2,py1,py2)
    penaltypoint=distanceval*distanceval
    if px1==px2 and py1!=py2 or py1 == py2 and px1!=px2:
        cost=distanceval
    elif px1 != px2 and py1 != py2:
        cost = distanceval+penaltypoint
    else:
        cost=distanceval
    return cost
```

```python
def find_path(graph, start_node):
    # Initialize path with starting node
    path = [start_node]

    # Continue searching as long as there's a node to process
    while path:
        # Get the last node in the current path
        current_node = path[-1]

        # If the current node has no further connections, break
        if not graph[current_node]:
            break

        # Find the node with the minimum weight connected to the current node
        min_node = min(graph[current_node], key=graph[current_node].get)

        # Append the found node to the path
        path.append(min_node)

        # Remove the found node from the graph to avoid revisiting
        graph[current_node].pop(min_node)

        print(path)

    return path
```
